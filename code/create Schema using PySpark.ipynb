```python
# Step 1: Import libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, lit, mean as _mean, expr
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, IntegerType

# Create Spark session
spark = SparkSession.builder.appName("Retail_Data_ETL").getOrCreate()

# Step 2: Create a sample retail dataset
data = [
    ("ORD001", "CUST001", "P001", "2024-12-01", 2, 250.0, "Electronics", "Store_A"),
    ("ORD002", "CUST002", "P002", "2024-12-01", 1, 120.0, "Clothing", "Store_B"),
    ("ORD003", "CUST003", "P003", "2024-12-02", 3, 75.5, "Grocery", "Store_A"),
    ("ORD004", "CUST002", "P004", "2024-12-03", 2, 500.0, "Electronics", "Store_C"),
    ("ORD005", "CUST004", "P005", "2024-12-03", 5, None, "Clothing", "Store_B"),
    ("ORD006", "CUST001", "P006", "2024-12-04", None, 90.0, "Grocery", "Store_A")
]

schema = StructType([
    StructField("OrderID", StringType(), True),
    StructField("CustomerID", StringType(), True),
    StructField("ProductID", StringType(), True),
    StructField("OrderDate", StringType(), True),
    StructField("Quantity", DoubleType(), True),
    StructField("UnitPrice", DoubleType(), True),
    StructField("Category", StringType(), True),
    StructField("Store", StringType(), True)
])

df = spark.createDataFrame(data, schema)
display(df)
```
